```python
hdb = pd.read_csv('hdb_lat_lon.csv')
```

```python
hdb.drop(columns=['Unnamed: 0', 'index'], inplace=True)
```

```python
hdb.head(5)
```

```python
import ast
```

```python
formatted_points = []

for i in hdb['lat_lon']:
    coords = ast.literal_eval(i)
    formatted_points.append({"latitude": coords[0], "longitude": coords[1]})

formatted_points
```

```python
for idx, i in enumerate(hdb['address']):
    formatted_points[idx]['address'] = i
```

```python
for idx, i in enumerate(hdb['address']):
    formatted_points[idx]['address'] = 'BLK ' + formatted_points[idx]['address']
```

```python
formatted_points[0]
```

```python
string_loc = []
for i in formatted_points:
    string_loc.append("{'latitude': '" + str(i['latitude']) + "', " + "'longitude': '" + str(i['longitude']) + "', " + "'address': '" + i['address'] + "'},")
```

```python
string_loc[0]
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(string_loc)
```

```python
df.to_csv('formatted_points.csv', index=False)
```
